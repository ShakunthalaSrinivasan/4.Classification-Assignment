```python
import pandas as pd
```

```python
dataset=pd.read_csv("CKD.csv")
dataset
```

```python
dataset=pd.get_dummies(dataset,drop_first=True,dtype='int')
dataset
```

```python
dataset.columns
```

```python
independent=dataset[['age', 'bp', 'al', 'su', 'bgr', 'bu', 'sc', 'sod', 'pot', 'hrmo', 'pcv',
       'wc', 'rc', 'sg_b', 'sg_c', 'sg_d', 'sg_e', 'rbc_normal', 'pc_normal',
       'pcc_present', 'ba_present', 'htn_yes', 'dm_yes', 'cad_yes',
       'appet_yes', 'pe_yes', 'ane_yes']]
dependent=dataset[['classification_yes']]
```

```python
dataset['classification_yes'].value_counts()
```

```python
from sklearn.model_selection import train_test_split
X_train,X_test,y_train,y_test=train_test_split(independent,dependent,test_size=0.30,random_state=0)
```

```python
from sklearn.preprocessing import StandardScaler
sc=StandardScaler()
X_train=sc.fit_transform(X_train)
X_test=sc.transform(X_test)
```

```python
from sklearn.model_selection import GridSearchCV
from sklearn.svm import SVC
param_grid={'kernel':['rbf','poly','linear','sigmoid'],'gamma':['scale','auto'],'C':[1,10,100,1000]}
grid=GridSearchCV(SVC(),param_grid,verbose=3,refit=True,n_jobs=-1,scoring='f1_weighted')
grid.fit(X_train,y_train)
```

```python
y_pred=grid.predict(X_test)
```

```python
result=grid.cv_results_
```

```python
table=pd.DataFrame.from_dict(result)
table
```

```python
print("The best model is {}:".format(grid.best_params_))
```

```python
from sklearn.metrics import confusion_matrix
cm=confusion_matrix(y_test,y_pred)
print(cm)
```

```python
from sklearn.metrics import classification_report
clf_report=classification_report(y_test,y_pred)
print(clf_report)
```

```python
import pickle
filename="finalised_model_SVC.sav"
pickle.dump(grid,open(filename,'wb'))
```
